```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import yfinance as yf
pd.set_option('display.max_rows', 10000)
plt.style.use('ggplot')
%matplotlib inline
```

```python
sp500russell2000 =list(pd.read_excel('sp500russell2000.xlsx')['Tickers'])
updated = pd.read_csv('swaggy_data.csv')[['tk','ment']]
updated = updated[updated['ment']>=5]
swaggy = list(updated['tk'])
```

```python
untreated = []
for sprussel in sp500russell2000:
    if sprussel not in swaggy:
        untreated.append(sprussel)
```

```python
from threading import Thread
data_list = []
issue = []
def process_data(stock):
    try:
        current_stock = yf.Ticker(stock)
        addition_list = []
        addition_list.append(stock)
        if stock in swaggy:
            addition_list.append(1)
        else:
            addition_list.append(0)
        info = current_stock.info
        addition_list.extend([info['shortName'], info['marketCap'], info['shortPercentOfFloat'], info['sector'], info['trailingEps']])
        nov = current_stock.history(start = '2021-01-05', end = '2021-01-06')
        addition_list.extend([float(nov['Close'][0])])
        jan = current_stock.history(start = '2021-02-03', end = '2021-02-04')
        addition_list.extend([float(jan['Close'][0])])              
        before = current_stock.history(start = '2021-02-04', end = '2021-02-05')
        addition_list.extend([float(before['Close'][0])])
        after = current_stock.history(start = '2021-02-18', end = '2021-02-19')
        addition_list.extend([float(after['Close'][0])])
        data_list.append(addition_list)
    except:
        issue.append(stock)
    print(stock)    
ticker_list = swaggy + untreated
for stock_thread in ticker_list:
    Thread(target=process_data, args=(stock_thread,)).start()
```

```python
stock_data = pd.DataFrame.from_records(data_list)
stock_data.columns = ['Symbol','Treated', 'CompanyName', 'MarketCap','ShortPercent', 'Sector', 'TrailingEPS', 'Nov2020', 'Jan2021', 'BeforeTreatment', 'AfterTreatment']
```

```python
stock_data.sort_values('Treated', ascending=False).to_csv('csds442_data2.csv')
```
